```python
df_junin = pd.read_csv("junin+girasol.csv", encoding='latin-1', sep=";")
df_junin
```

```python
fig, ax = plt.subplots(figsize=(7,5))  
mapeo_colores = ax.scatter(df_junin['Campaña'], df_junin['Rendimiento'], c=df_junin['Precipitaciones'])
fig.colorbar(mapeo_colores)
ax.set_title('Relacion Rendimiento/Precipitaciones en temporada agricola')
ax.set_xlabel('Campaña')
ax.set_ylabel('Rendimiento')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7,5))  
mapeo_colores = ax.scatter(df_junin['Sup. Sembrada'], df_junin['Rendimiento'], c=df_junin['Precipitaciones'])
fig.colorbar(mapeo_colores)
ax.set_title('Relacion Rendimiento/Precipitaciones en temporada agricola')
ax.set_xlabel('Superficie Sembrada')
ax.set_ylabel('Rendimiento')

plt.show()
```
